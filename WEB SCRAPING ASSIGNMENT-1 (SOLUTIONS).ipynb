# 2) Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. name, rating, year of release) and make data frame.

```python
#Solution

import pandas as pd
import requests
from bs4 import BeautifulSoup

page2=requests.get('https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc&count=100')
#page2
soup2 = BeautifulSoup(page2.content, "html.parser")

Movie_name = []
Rating = []
Year_of_release = []

movie_data = soup2.findAll('div', attrs={'class':'lister-item mode-advanced'})

for i in movie_data:
    name = i.h3.a.text
    Movie_name.append(name)
    
    rate = i.find('div', class_= 'inline-block ratings-imdb-rating').text.replace('\n', '')
    Rating.append(rate)
    
    year = i.h3.find('span', class_= 'lister-item-year text-muted unbold').text.replace ('(','').replace(')','')
    Year_of_release.append(year)
    
       
df_movie=pd.DataFrame({'Movie name':Movie_name, 'Rating': Rating, 'Year of release':Year_of_release})
df_movie
```

# 3) Write a python program to display IMDB’s Top rated 100 Indian movies’ data (i.e. name, rating, year of release) and make data frame.

```python
#Solution

import pandas as pd
import requests
from bs4 import BeautifulSoup

page3 = requests.get('https://www.imdb.com/list/ls009997493/')
#page3
soup3 = BeautifulSoup(page3.content, "html.parser")

Movie_name = []
Rating = []
Year_of_release = []

movie_data = soup3.findAll('div', attrs={'class':'lister-item-content'})

for store in movie_data:
    name = store.h3.a.text
    Movie_name.append(name)
    
    rate = store.find('div', class_= 'ipl-rating-star small').text.replace('\n', '')
    Rating.append(rate)
    
    year = store.h3.find('span', class_= 'lister-item-year text-muted unbold').text.replace ('(','').replace(')','')
    Year_of_release.append(year)
    
df_indian_movie=pd.DataFrame({'Movie name':Movie_name, 'Rating': Rating, 'Year of release':Year_of_release})
df_indian_movie
```

# 4) Write a python program to scrape cricket rankings from icc-cricket.com. You have to scrape:
a) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating.
b) Top 10 ODI Batsmen in men along with the records of their team and rating.
c) Top 10 ODI bowlers along with the records of their team and rating.

```python
#Solution

import requests
from bs4 import BeautifulSoup
 
print('---------------------------------------------------------------------------------')
print('(a)Top 10 ODI teams in men’s cricket along with the records for matches, points and rating.')
print('---------------------------------------------------------------------------------')
page4a=requests.get('https://www.icc-cricket.com/rankings/mens/team-rankings/odi')
#page4a
soup4a = BeautifulSoup(page4a.content, "html.parser")     
 
for tr in soup4a.find_all('tr'):
    tds = []
    for th in tr.find_all('th'):
        tds.append(th.text.replace('\n',' '))
        
    for td in tr.find_all('td'):
        tds.append(td.text.replace('\n',' ').strip())
        
    print(tds)
    

print('---------------------------------------------------------------------------------')
print('(b) Top 10 ODI Batsmen in men along with the records of their team and rating.')
print('---------------------------------------------------------------------------------')
page4b=requests.get('https://www.icc-cricket.com/rankings/mens/player-rankings/odi/batting')
#page4b
soup4b = BeautifulSoup(page4b.content, "html.parser") 
 
for tr in soup4b.find_all('tr'):
    mds = []
    for th in tr.find_all('th'):
        mds.append(th.text.replace('\n',' '))
        
    for td in tr.find_all('td'):
        mds.append(td.text.replace('(0)',' ').strip())
        
    print(mds)
    
       
print('---------------------------------------------------------------------------------')
print('(c) Top 10 ODI bowlers along with the records of their team and rating.')
print('---------------------------------------------------------------------------------')
page4c=requests.get('https://www.icc-cricket.com/rankings/mens/player-rankings/odi/bowling')
#page4c
soup4c = BeautifulSoup(page4c.content, "html.parser")  
 
for tr in soup4c.find_all('tr'):
    alls = []
    for th in tr.find_all('th'):
        alls.append(th.text.replace('\n',' '))
        
    for td in tr.find_all('td'):
        alls.append(td.text.replace('(0)',' ').strip())
        
    print(alls)
```

# 6) Write a python program to scrape details of all the posts from coreyms.com. Scrape the heading, date, content and the code for the video from the link for the youtube video from the post.

```python
#Solution:

import requests
from bs4 import BeautifulSoup
 
url = 'https://coreyms.com/'
page6=requests.get(url)
#page6
soup6 = BeautifulSoup(page6.content, "html.parser")

print('---------------------------')
print('Heading:')
print('---------------------------')    
heading = soup6.find_all('h2', attrs={'class':'entry-title'})
for hd in heading:
    print(hd.get_text())
    

print('---------------------------')
print('Date:')
print('---------------------------')    
date = soup6.find_all('time', attrs={'class':'entry-time'})
for dt in date:
    print(dt.get_text())
    
    
print('---------------------------')
print('Content:')
print('---------------------------')    
content = soup6.find_all('div', attrs={'class':'entry-content'})
for ct in content:
    print(ct.get_text())
    

print('---------------------------')
print('Code for the video from the link for the youtube video:')
print('---------------------------')
links = soup6.find_all('iframe', attrs={'class':'youtube-player'})
for link in links:
    print(link['src'])
```

# 7) Write a python program to scrape house details from mentioned URL. It should include house title, location,area, EMI and price from nobroker.in.

```python
#Solution

import requests
from bs4 import BeautifulSoup
import pandas as pd
 
url = 'https://www.nobroker.in/flats-for-sale-in-koramangala_bangalore'
page7=requests.get(url)
#page7
soup7 = BeautifulSoup(page7.content, "html.parser")
 
    
print('---------------------------')
print('House title:')
print('---------------------------')    
h_title = soup7.find_all('h2', attrs={'class':'heading-6 font-semi-bold nb__25Cl7'})
for ht in h_title:
    print(ht.get_text())
    
    
print('---------------------------')
print('Location:')
print('---------------------------')    
h_loc = soup7.find_all('div', attrs={'class':'nb__1EwQz'})
for hl in h_loc:
    print(hl.get_text())
    
    
print('---------------------------')
print('Area:')
print('---------------------------')    
h_area = soup7.find_all('div', attrs={'class':'nb__FfHqA'})
for ha in h_area:
    print(ha.get_text())
    
    
print('---------------------------')
print('EMI:')
print('---------------------------')    
h_emi = soup7.find_all(id="roomType", attrs={'class':'font-semi-bold heading-6'})
for emi in h_emi:
    print(emi.get_text())
    
    
print('---------------------------')
print('Price:')
print('---------------------------')    
h_price = soup7.find_all(id="minDeposit", attrs={'class':'nb__4L90a'})
for price in h_price:
    print(price.get_text())
    

df_house=pd.DataFrame({'House title':h_title, 'Location': h_loc, 'Area':h_area, 'EMI':h_emi, 'Price':h_price})
df_house
```

# 8) Write a python program to scrape mentioned details from dineout.co.in :     (i) Restaurant name (ii) Cuisine (iii) Location (iv) Ratings (v) Image URL

```python
#Solution

import pandas as pd
import requests
from bs4 import BeautifulSoup
 
url = 'https://www.dineout.co.in/delhi-restaurants/'
page8=requests.get(url)
#page8 
soup8 = BeautifulSoup(page8.content, "html.parser")
 
    
print('---------------------------')
print('(i) Restaurant name:')
print('---------------------------')    
res = soup8.find_all('div', attrs={'class':'restnt-info cursor'})
for rn in res:
    print(rn.get_text())
 

print('---------------------------')
print('(ii) Cuisine:')
print('---------------------------')
cns = soup8.find_all('span', class_="double-line-ellipsis") 
for c in cns:
    print(c.text.split('|')[1])
    
           
print('---------------------------')
print('(iii) Location:')
print('---------------------------')    
loc = soup8.find_all('div', attrs={'class':'restnt-loc ellipsis'})
for l in loc:
    print(l.get_text())

    
print('---------------------------')
print('(iv) Ratings:')
print('---------------------------')    
rate = soup8.find_all('div', attrs={'class':'restnt-rating rating-4'})
for rt in rate:
    print(rt.get_text())
    
    
      
print('---------------------------')
print('image URL')
print('---------------------------')
pics = soup8.find_all('img',attrs={'class':'no-img'})
for img in pics:
    print(img['data-src']) 
    
    

df_Restaurant=pd.DataFrame({'Restaurant name':res, 'Cuisine':cns, 'Location':loc, 'Ratings':rate, 'image URL':pics})
df_Restaurant
```
